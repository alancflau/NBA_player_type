```python
def stats_retrieval(year, interest):
    
    """
    To retrieve data for specific field
    
    Params:
    year: Year of date of interest
    interest: particular field we want to search: 
        advanced stats, per 100 poss stats, shooting stats
    """
    
    url = 'https://www.basketball-reference.com/leagues/NBA_{}_{}.html'.format(year, interest)
    df = pd.read_html(url)[0]
    
    # CASE
    if interest = 'shooting':
        df.columns = df[0].columns.get_level_values(0)
        
        # TODO: Rename columns
        ['% of FGA by Distance', 'FG% by Distance', '% of FG Ast']
        # For more information
        https://stackoverflow.com/questions/24290297/pandas-dataframe-with-multiindex-column-merge-levels
    
    
    # Remove columns that contains Unnamed
    if any(s.startswith('Unnamed') for s in df.columns):
        return df.loc[:, ~df.columns.str.contains('^Unnamed')]
    
    else:
        return df
```

```python
advanced_dfs = []
per_100_poss_dfs = []
shooting_dfs = []

for year in tqdm(range(2015,2021)):
    advanced_df = stats_retrieval(year, 'advanced')
    advanced_dfs.append(advanced_df)
    
    per_100_poss_df = stats_retrieval(year, 'per_poss')
    per_100_poss_dfs.append(per_100_poss_df)    
    
    shooting_df = stats_retrieval(year, 'shooting')
    shooting_dfs.append(shooting_df)    
```

```python
pd.concat(advanced_dfs)
```
